```python
import yfinance as yf

df = yf.Ticker('BAC')
df_cf = df.cash_flow
df_cf
```

```python
df_bs = df.balance_sheet
df_bs
```

```python
df_is = df.incomestmt
df_is
```

```python
net_inc = df_is.loc['Net Income']
int_exp = df_is.loc['Interest Expense']
dep_amo = df_cf.loc['Depreciation And Amortization']
tax = df_is.loc['Tax Provision']
tot_rev = df_is.loc['Total Revenue']
```

```python
import pandas as pd
big_df = pd.concat([net_inc, int_exp, dep_amo, tax], axis=1)
```

```python
big_df['EBITDA'] = net_inc+int_exp+dep_amo+tax
big_df['EBITDA Margin'] = (net_inc+int_exp+dep_amo+tax)/ tot_rev
big_df
```
